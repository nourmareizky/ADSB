```python
x1 = imdb[imdb.imdb_score>=7]
x2 = imdb[imdb.imdb_score<7]
x1
```

```python
Dir_x1 = x1[x1.director_top20 == 1]
Dir_xx1 = x1[x1.director_top20 == 0]
Dir_x2 = x2[x2.director_top20 == 1]
print(len(Dir_x1)/len(x1)*100)
print(len(Dir_x2)/len(x2)*100)
```

```python
Act1_x1 = x1[x1['actor_top20.1'] == 1]
Act1_xx1 = x1[x1['actor_top20.1'] == 0] 
Act1_x2 = x2[x2['actor_top20.1'] == 1]
print(len(Act1_x1)/len(x1))
print(len(Act1_x2)/len(x2))
```

```python
Act2_x1 = x1[x1['actor_top20.2'] == 1]
Act2_xx1 = x1[x1['actor_top20.2'] == 0]
Act2_x2 = x2[x2['actor_top20.2'] == 1]
print(len(Act2_x1)/len(x1))
print(len(Act2_x2)/len(x2))
```
