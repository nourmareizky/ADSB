# SEABORN
anak matplotlib. Kalau import sns harus import matplotlib. 
seaborn bisa sekali panggil tdk sebanyak matplotlib

```python
tips = sns.load_dataset('tips') #bisa ambil data dari semua dataset di https://github.com/mwaskom/seaborn-data
tips.head()
```

```python
sns.lmplot(x='total_bill', y = 'tip', data=tips, hue='sex',palette='Set1') #linear model plot. Linear model + regresi
plt.show()
```

```python
sns.lmplot(x='total_bill', y = 'tip', data=tips, hue='day',col = 'sex') #linear model plot. Linear model + regresi
plt.show()
```

```python
sns.stripplot(x='day',y='tip', data=tips, size=8, jitter=True) #jitter untuk mecah garis
plt.ylabel('tip($)')
plt.title('Tip Distribution')
plt.show()
```

1 variabel -> strippolt. Untuk melihat distribusi dr tip. paling banyak kasih tip sekitar 1-5. Ada outliers di tip 10$

```python
sns.swarmplot(x = 'day',y='tip',data = tips,hue='sex') #orient = 'h' -> orientasi horizontal
plt.ylabel('tip($)')
plt.title('Tip Distribution')
plt.show()
```

```python
sns.swarmplot(x = 'tip',y='day',data = tips,hue='sex',orient='h') #orient = 'h' -> orientasi horizontal
plt.ylabel('tip($)')
plt.axis('tight')
plt.title('Tip Distribution')
plt.show()
```

```python
sns.boxplot(x = 'day',y='tip',data = tips) #orient = 'h' -> orientasi horizontal
plt.ylabel('tip($)')
#plt.axis('tight')
plt.tight_layout()
plt.title('Tip Distribution')
plt.show()
```

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
plt.subplots(figsize=(15,4))
sns.stripplot(x='month',y='passengers', data=flight, size=8,hue='year')#jitter untuk mecah garis
plt.ylabel('')
plt.legend(bbox_to_anchor=(1,1)) #legend outside box
#plt.title('Tip Distribution')
sns.despine()
plt.show()
```

```python
sns.stripplot(x='passengers',y='year', data=flight, size=8, jitter=True,orient='h') #jitter untuk mecah garis
plt.ylabel('year')
#plt.title('Tip Distribution')
plt.show()
```

```python
sns.lmplot(x='month', y = 'passengers', data=flight,palette='Set1') #linear model plot. Linear model + regresi
plt.show()
```

### Heatmap

```python
from sklearn import preprocessing
from matplotlib.pyplot import cm
```

```python
std_scale = preprocessing.StandardScaler().fit(tips[['total_bill','tip','size']]) #mengambil dan membentuk dataframe baru
df_std = std_scale.transform(tips[['total_bill','tip','size']])

df = pd.DataFrame(df_std)
df.head()
```

```python
print(df.cov())
sns.heatmap(df.cov(),cmap=cm.Blues, annot=True)
plt.show()
```
