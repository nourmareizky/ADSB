```python
auto = pd.read_csv('imports-85.data.txt',index_col=None)
auto.head(20)

```

```python
auto.info()
```

```python
body_style=auto.groupby('body-style')['symboling'].count()
```

```python
def cleanUp(column):
    auto[column]=pd.to_numeric(auto[column], errors='coerce').astype(np.float)
    auto[column]=auto[column].fillna(auto[column].mean())
auto.head()
list_value=['normalized-losses','symboling','peak-rpm','city-mpg','horsepower','highway-mpg','price']

for n in list_value:
    cleanUp(n)
```
