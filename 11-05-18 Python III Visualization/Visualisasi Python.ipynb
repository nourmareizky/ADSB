```python
#Seaborn (child matplotlib for new styling and automate)
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips) #scaterplot to regression
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='Set1') #scaterplot to regression
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='day', col='sex')
plt.show()
```

```python
#strip plot
sns.stripplot(y='tip', data = tips)
plt.ylabel('tip ($)')
plt.title("Tip Distribusi")
plt.show()
```

```python
sns.stripplot(y='tip', data = tips, jitter = True)
plt.ylabel('tip ($)')
plt.title("Tip Distribusi")
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data = tips, jitter = True)
plt.xlabel('Day')
plt.ylabel('tip ($)')
plt.title("Tip Distribusi")
plt.show()
```

```python
sns.swarmplot(x='sex', y='tip', data = tips)
plt.xlabel('Sex')
plt.ylabel('tip ($)')
plt.title("Tip Distribusi")
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data = tips, hue='sex')
plt.xlabel('Day')
plt.ylabel('tip ($)')
plt.title("Tip Distribusi")
plt.show()
```

```python
sns.swarmplot(x='tip', y='day', data = tips, hue='sex', orient='h')
plt.xlabel('Day')
plt.ylabel('tip ($)')

plt.axis('tight')
plt.title("Tip Distribusi")
plt.show()
```

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')
plt.tight_layout()
plt.show()
```

```python
#join plot
sns.jointplot(x= 'total_bill', y='tip', data= tips, kind='kde')
plt.show()
```

```python
#hexa plot
sns.jointplot(x= 'total_bill', y='tip', data= tips, kind='hex')
plt.show()
```

```python
#Pair Plot
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
#Heatmap
from sklearn import preprocessing #machine learning library
from matplotlib.pyplot import cm
std_scale = preprocessing.StandardScaler().fit(tips[['total_bill','tip','size']])
df_std = std_scale.transform(tips[['total_bill','tip','size']])
df = pd.DataFrame(df_std)
df.head()
```

```python
print(df.cov())
sns.heatmap(df.cov(),cmap=cm.Blues, annot=True)
plt.show()
```
