```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.toyota.astra.co.id/toyota-connect/news/modifikasi-toyota-alphard-hercules-terlihat-horor'

# Make the request and create the response object: response
response = requests.get(url)

# Extract HTML texts contained in Response object: html
html = response.text
```

```python
print (html)
```

```python
# Create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(html, "html5lib")
```

```python
# Make it to show nested structure
print (soup.prettify())
```

```python
soup_txts = soup.find_all('p', style="text-align: justify;")
```

```python
print (soup_txts[3].text)
```

```python
text = ''

for txt in soup_txts:
    text = text + txt.text
```
