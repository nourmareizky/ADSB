```python
import pandas as pd
import numpy as np
```

```python
owner = pd.read_csv("Owners.csv")
```

```python
pets1 = pd.read_excel("Pets1.xlsx")
pets2 = pd.read_excel("Pets2.xlsx")
```

```python
owner.shape
```

```python
pet = pd.concat([pets1,pets2]).reset_index(drop=True)
```

```python
pet.shape
```

```python
pet
```

```python
#merge pet and owner
data = pd.merge(pet,owner, how='outer', on='OwnerID')

#drop duplicates
data.drop_duplicates(inplace = True)

```

```python
data.isnull().sum()
```

```python

data[data.Kind =='Dog'] = data.fillna(value={'Gender':'male'})
data[data.Kind =='Cat'] = data.fillna(value={'Gender':'female'})
data[data.Kind =='Parrot'] = data.fillna(value={'Gender':'male'})

rata2 = ((data.groupby(["Gender","Kind"])).mean())
rata2
#data = data[data['Kind']=='Parrot'].fillna(value={'Gender':'male'})
```

```python
#data["Age"].fillna(data.groupby(["Gender","Kind"])["Age"].transform("mean"), inplace=True)

#data

```

```python
rata2 = ((data.groupby(["Gender","Kind"])).mean())['Age']
rata2
```

```python
data[(data.Kind == 'Cat') & (data.Gender == 'female')] = data.fillna(value = rata2[0])
data[(data.Kind == 'Dog') & (data.Gender == 'female')] = data.fillna(value = rata2[1])
data[(data.Kind == 'Parrot') & (data.Gender == 'female')] = data.fillna(value = rata2[2])
data[(data.Kind == 'Cat') & (data.Gender == 'male')] = data.fillna(value = rata2[3])
data[(data.Kind == 'Dog') & (data.Gender == 'male')] = data.fillna(value = rata2[4])
data[(data.Kind == 'Parrot') & (data.Gender == 'male')] = data.fillna(value = rata2[5])
```
